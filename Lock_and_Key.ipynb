```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Simulation parameters
num_keys = 10
num_people = 10000

# Simulate the experiments
results = []
for _ in range(num_people):
    keys = list(range(num_keys))  # Create a list of keys
    np.random.shuffle(keys)      # Shuffle the keys randomly
    correct_key = keys[0]        # Assume the first key is the correct one after shuffling

    # Count the attempts until the correct key is found
    '''for i, key in enumerate(keys, start=1):
        if key == 0:  # Correct key is labeled as 0
            results.append(i)
            break'''
    i=1
    while True:
      v = np.random.randint(0, 10)
      if v==correct_key:
        results.append(i)
        break
      i+=1

# Count the frequency of each k
k_counts = Counter(results)
k_values, frequencies = zip(*k_counts.items())

# Visualize the distribution
plt.figure(figsize=(10, 6))
plt.bar(k_values, frequencies, color='skyblue', edgecolor='black')
plt.xlabel('Attempt Number (k)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Distribution of Attempts to Open the Lock', fontsize=16)
plt.xticks(range(1, num_keys + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Find the most likely value of k
most_likely_k = max(k_counts, key=k_counts.get)
print(f"The most likely value of k is: {most_likely_k}")
```
